```python
import pandas as pd
```

```python
df = pd.read_csv('../data/cleaned_dataset_step_3.csv')
```

```python
df.isnull().sum()
```

```python
city_country_mapping = pd.read_csv('../data/worldcities.csv')
```

```python
df['city'] = df['Country'].apply(lambda x: x.split(',')[0].strip() if isinstance(x, str) else None)
```

```python
print(df['city'].unique())
```

```python
print("Unique cities in df:", df['city'].unique())
print("Unique cities in city_country_mapping:", city_country_mapping['city'].unique())
```

```python
# Strip spaces and convert to lowercase
df['city'] = df['city'].str.strip().str.lower()
city_country_mapping['city'] = city_country_mapping['city'].str.strip().str.lower()
print(df['city'])
print(city_country_mapping['city'])
```

```python
city_country_mapping = city_country_mapping.drop_duplicates(subset=['city'], keep='first')
```

```python
df = df.merge(city_country_mapping, how='left', on='city')
print("Merge result (head):")
print(df.head())
```

```python
unmatched_cities = df[~df['city'].isin(city_country_mapping['city'])]
print("Unmatched cities:")
print(unmatched_cities['city'].unique())
print(len(unmatched_cities['city'].unique()))
```

```python
df
```

```python
df = df.drop(columns=['city','city_ascii',	'lat',	'lng',	'iso2',	'iso3'	,'admin_name',	'capital',	'population',	'id'])
```

```python
df
```

```python
df["Country"] = df["country"].combine_first(df["Country"])
```

```python
df.drop(columns=['country'],inplace=True)
```

```python
df.to_csv('../data/cleaned_dataset_step_4.csv', index=False)
print("Results saved to '('../data/cleaned_dataset_step_4.csv'.")
```
